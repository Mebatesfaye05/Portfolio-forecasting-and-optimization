```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


aapl = yf.Ticker("AAPL")
aapl_data = aapl.history(start='2020-01-01', end='2022-12-31', auto_adjust=True)

googl = yf.Ticker("GOOGL")
googl_data = googl.history(start='2020-01-01', end='2022-12-31', auto_adjust=True)

msft = yf.Ticker("MSFT")
msft_data = msft.history(start='2020-01-01', end='2022-12-31', auto_adjust=True)
```

```python
##Combine the data into a single data frame using pandas.

data = pd.concat([aapl_data, googl_data, msft_data], keys=['AAPL', 'GOOGL', 'MSFT'], names=['Company'])
data

```

```python
##merge the data to single dataframe
data = pd.concat([aapl_data, googl_data, msft_data], keys=['AAPL', 'GOOGL', 'MSFT'], names=['Company', 'Date'])
data
```

```python
##merge the data to single dataframe
data = pd.concat([aapl_data, googl_data, msft_data], keys=['AAPL', 'GOOGL', 'MSFT'], names=['Company', 'Date'])
data
```

```python
pivot_table = pd.pivot_table(data, values=['Open', 'High', 'Low', 'Close'], index='Date', columns='Company')
pivot_table


```

```python
#Flatten the pivot table to create a single data frame with all the data in a long format.
flat_table = pivot_table.stack().reset_index().rename(columns={'level_1': 'Metric', 0: 'Value'})
flat_table

```

```python
# Calculate daily returns
returns = flat_table.groupby('Company')['Close'].pct_change()

# Add the daily returns to the flat table
flat_table['Return'] = returns

# Print the updated flat table with the daily returns
display(flat_table)

```

```python
# Calculate summary statistics for the daily returns
display(flat_table.groupby('Company')['Return'].describe())
```

```python
# Plot the daily returns for each company
sns.set_style('darkgrid')
plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='Return', hue='Company', data=flat_table)
plt.title('Daily Returns for AAPL, GOOGL, and MSFT')
plt.xlabel('Date')
plt.ylabel('Return')
plt.show()
```

```python
# Calculate correlations between the daily returns for each pair of companies
corr = flat_table.pivot_table(values='Return', index='Date', columns='Company').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix for Daily Returns')
plt.show()
```

```python
# Calculate the expected return for each company 
expected_return = flat_table.groupby('Company')['Return'].mean() * 252
print(expected_return)

# Plot the expected return for each company
plt.bar(expected_return.index, expected_return.values)
plt.xticks(rotation=45)
plt.ylabel('Expected Return')
plt.title('Expected Return for Each Company')
plt.show()
```

```python
# Calculate the volatility for each company based on the historical data in the flat_table DataFrame
volatility = flat_table.groupby('Company')['Return'].std() * np.sqrt(252)
volatility
```

```python
# Define the risk-free rate
risk_free_rate = 0.02
```

```python
# Calculate the Sharpe ratio for each company
sharpe_ratio = (expected_return - risk_free_rate) / volatility
display(sharpe_ratio)
```

```python
# Create a bar plot of the Sharpe ratio for each company
plt.bar(sharpe_ratio.index, sharpe_ratio.values)
plt.xticks(rotation=45)
plt.axhline(y=0, color='black', linestyle='--')
plt.ylabel('Sharpe Ratio')
plt.title('Sharpe Ratio for AAPL, GOOGL, and MSFT')
plt.show()

```

```python

# Calculate the Sortino ratio for each company



# Calculate the downside deviation for each company based on the historical data in the flat_table DataFrame
downside_deviation = flat_table.groupby('Company')['Return'].apply(lambda x: np.std(x[x < 0])) * np.sqrt(252)

# Calculate the Sortino ratio for each company
sortino_ratio = (expected_return - risk_free_rate) / downside_deviation

# Print the Sortino ratio for each company
sortino_ratio
```

```python
import numpy as np
import random


# Define the number of portfolios to simulate
num_portfolios = 10000

# Generate random portfolio weights
weights_list = []
for i in range(num_portfolios):
    weights = np.array([random.random() for i in range(len(expected_return))])
    weights /= sum(weights)
    weights_list.append(weights)
```

```python
# Calculate the expected return and volatility of each portfolio
port_returns = []
port_volatility = []
for weights in weights_list:
    port_returns.append(sum(weights * expected_return))
    port_volatility.append(np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights))))
```

```python
# Convert the lists to arrays
port_returns = np.array(port_returns)
port_volatility = np.array(port_volatility)
```

```python

# Plot the efficient frontier
plt.figure(figsize=(10, 6))
plt.scatter(port_volatility, port_returns, c=port_returns / port_volatility, cmap='coolwarm')
plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Volatility')
plt.ylabel('Expected Return')
plt.title('Efficient Frontier')
plt.show()
```
